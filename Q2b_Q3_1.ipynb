## 1 Sample KS Test

```python
# 1-Sample KS Test for "cases"

# Poisson Distribution
def poisson_mme(X):
    return np.mean(X)

# Geometric Distribution
def geometric_mme(X):
    return 1.0/np.mean(X)

def binomial_mme(X):
    mu = np.mean(X)
    var = np.var(X)
    n_hat = (mu**2)/(mu-var)
    p_hat = mu/n_hat 
    return n_hat, p_hat

#Parameters from last 3 months of 'cases' from MN
X = df_mn_last_3_months['cases'].to_numpy()
poisson_lambda = poisson_mme(X)
geometric_p =  geometric_mme(X)
binomial_n, binomial_p = binomial_mme(X)
```

```python
print(f"Poisson lambda Guess: {poisson_lambda}")
print(f"Geometric_p Guess: {geometric_p}")
print(f"binomial_n Guess: {binomial_n} binomial_p Guess:{binomial_p}")
```

```python
# 1-Sample KS Test for Poisson

from scipy.stats import poisson
from scipy.stats import uniform
from collections import Counter

# def eCDF(D):
#     n = len(D)
#     f_hat_pos = []
#     f_hat_neg = []
#     for i,x in enumerate(D):
#         f_hat_pos.append((i+1)/n)
#         f_hat_neg.append(i/n)
#     return np.array(f_hat_neg), np.array(f_hat_pos)

def eCDF(D):
    n = len(D)
    f_hat_pos = []
    f_hat_neg = []

    counts = Counter(D)
    
    neg=0
    pos=counts[D[0]]/n
    f_hat_pos.append(pos)
    f_hat_neg.append(neg)
    X = list(counts.keys())
    X.sort()
    for i,x in enumerate(X[1:]):
        neg = pos
        pos = pos + (counts[x]/n)
        f_hat_pos.append(pos)
        f_hat_neg.append(neg)
        
    return np.array(f_hat_neg), np.array(f_hat_pos)
    
    
# eCDF([0,0,0,1,1,2])
    for i,x in enumerate(D):
        f_hat_pos.append((i+1)/n)
        f_hat_neg.append(i/n)
    return np.array(f_hat_neg), np.array(f_hat_pos)

def poisson_cdf(X, parameter):
    return np.array(poisson.cdf(X, parameter))

def KS_Test_1_sample(X, e_cdf, target_cdf,parameter):
    X = list(Counter(X).keys())
    X.sort()
    f_hat_neg, f_hat_pos = e_cdf(X)
    f = target_cdf(X, parameter)
    return max(max(abs(f-f_hat_neg)), max(abs(f-f_hat_pos)))
    

D = df_mt_last_3_months['cases'].to_numpy()
KS_statistic = KS_Test_1_sample(D, eCDF, poisson_cdf , poisson_lambda)
print(f"H0: CDF of MT Cases = CDF of Poisson ({poisson_lambda})")
print(f"H1: CDF of MT Cases != CDF of Poisson ({poisson_lambda})")
print(f"Since {KS_statistic}(KS Statistic) > 0.05(Critical Value), we reject H0.")
```

```python
# 1-Sample KS Test for Geometric
from scipy.stats import geom
def geometric_cdf(X, parameter):
    return np.array(geom.cdf(X, parameter))

KS_statistic = KS_Test_1_sample(D, eCDF, geometric_cdf , geometric_p)
print(f"H0: CDF of MT Cases = CDF of Geometric ({geometric_p})")
print(f"H1: CDF of MT Cases != CDF of Geometric ({geometric_p})")
print(f"Since {KS_statistic}(KS Statistic) > 0.05(Critical Value), we reject H0.")
```

```python
# 1-Sample KS Test for Binomial
from scipy.stats import binom

def binomial_cdf(X, parameters):
    return np.array(binom.cdf(X, parameters['n'], parameters['p']))

KS_statistic = KS_Test_1_sample(D, eCDF, binomial_cdf , {'n':binomial_n, 'p':binomial_p})
print(f"H0: CDF of MT Cases = CDF of Binomial ({binomial_n, binomial_p})")
print(f"H1: CDF of MT Cases != CDF of Binomial ({binomial_n, binomial_p})")
print(f"Since {KS_statistic}(KS Statistic) > 0.05(Critical Value), we reject H0.")
```

```python
def get_KS_statistic(X,D,c,field):
    
    poisson_lambda = poisson_mme(X)
    geometric_p =  geometric_mme(X)
    binomial_n, binomial_p = binomial_mme(X)
    
    KS_statistic = KS_Test_1_sample(D, eCDF, poisson_cdf , poisson_lambda)
    print("Poisson Distribution")
    print(f"H0: CDF of MT {field} = CDF of Poisson Distribution with parameter ({poisson_lambda})")
    print(f"H1: CDF of MT {field} != CDF of Poisson Distribution with parameter ({poisson_lambda})")
    if KS_statistic > c:
        print(f"Since {KS_statistic}(KS Statistic) > {c}(Critical Value), we reject H0.\n")
    else:
        print(f"Since {KS_statistic}(KS Statistic) <= {c}(Critical Value), we accept H0.\n")
    
    KS_statistic = KS_Test_1_sample(D, eCDF, geometric_cdf , geometric_p)
    print("Geometric Distribution")
    print(f"H0: CDF of MT {field} = CDF of Geometric Distribution with parameter ({geometric_p})")
    print(f"H1: CDF of MT {field} != CDF of Geometric Distribution with parameter ({geometric_p})")
    if KS_statistic > c:
        print(f"Since {KS_statistic}(KS Statistic) > {c}(Critical Value), we reject H0.\n")
    else:
        print(f"Since {KS_statistic}(KS Statistic) <= {c}(Critical Value), we accept H0.\n")
    
    KS_statistic = KS_Test_1_sample(D, eCDF, binomial_cdf , {'n':binomial_n, 'p':binomial_p})
    print("Binomial Distribution")
    print(f"H0: CDF of MT {field} = CDF of Binomial Distribution with parameter ({binomial_n, binomial_p})")
    print(f"H1: CDF of MT {field} != CDF of Binomial Distribution with parameter ({binomial_n, binomial_p})")
    if KS_statistic > c:
        print(f"Since {KS_statistic}(KS Statistic) > {c}(Critical Value), we reject H0.\n")
    else:
        print(f"Since {KS_statistic}(KS Statistic) <= {c}(Critical Value), we accept H0.\n")

for field in ['cases', 'deaths']:
    print(f"1-sample KS Test for {field}\n")
    X = df_mn_last_3_months[field].to_numpy()
    D = df_mt_last_3_months[field].to_numpy()
    get_KS_statistic(X, D, 0.05, field)
    print("\n\n")
```

```python
def get_D2_eCDF(D1,D2):
    def firstGreaterThanEqualTo(data, val):
        for i,x in enumerate(data):
            if x>= val:
                return i
        return len(data)

    d2_neg, d2_pos = eCDF(D2)
    x_neg = []
    x_pos = []
    D2 = list(Counter(D2).keys())
    D2.sort()
    for x in np.unique(D1):
        idx = firstGreaterThanEqualTo(D2,x)
        if idx == len(D2):
            x_neg.append(1.0)
            x_pos.append(1.0)
        else:
            if x == D2[idx]:
                x_pos.append(d2_pos[idx])
                x_neg.append(d2_neg[idx])
            else:
                x_neg.append(d2_neg[idx])
                x_pos.append(d2_neg[idx])
    return np.array(x_neg), np.array(x_pos)
```

## 2- Sample KS Test

### D1 = Oct-Dec Data of 1st State MN, D2 = Oct-Dec Data of 2nd State MT

```python
def KS_test_2_sample(D1, D2):
    D1.sort()
    D2.sort()
    
    d1_neg, d1_pos = eCDF(D1)
    d2_neg, d2_pos = get_D2_eCDF(D1, D2)
    
    return max(max(abs(d1_neg-d2_neg)), max(abs(d1_pos - d2_pos)))

for field in ['cases','deaths']:
    print(f"2 sample KS Test for field {field}\n")
    print(f"H0: CDF of MN {field} = CDF of MT {field}")
    print(f"H1: CDF of MN {field} != CDF of MT {field}")
    D1 = df_mn_last_3_months[field].to_numpy()
    D2 = df_mt_last_3_months[field].to_numpy()
    KS_Statistic = KS_test_2_sample(D1, D2)
    if KS_Statistic > 0.05:
        print(f"{KS_Statistic} > 0.05, therefore reject H0\n")
    else:
        print(f"{KS_Statistic} <= 0.05, therefore accept H0\n")
```

## Permutation Test
### D1 = Oct-Dec Data of 1st State MN, D2 = Oct-Dec Data of 2nd State MT

```python
# Permutation Test

def get_permutations(D):
    permutation_set = set()
    permutations = []
    while len(permutations)<1000:
        perm = np.random.permutation(D)
        if tuple(perm) not in permutation_set:
            permutations.append(perm)
    return permutations

for field in ['cases','deaths']:
    D1 = df_mn_last_3_months[field].to_numpy()
    D2 = df_mt_last_3_months[field].to_numpy()

    n = len(D1) + len(D2)
    D = np.array(D1.tolist()+D2.tolist())

    t_obs = abs(np.mean(D1)-np.mean(D2))

    perms = get_permutations(D)
    count = 0

    for perm in perms:
        l = np.array(list(perm))
        d1 = l[:len(D1)]
        d2 = l[len(D1):]
        t = abs(np.mean(d1)-np.mean(d2))
        if t > t_obs:
            count += 1
    p_value = count/len(perms)
    print(f"H0: MN {field} = MT {field}")
    print(f"H1: MN {field} != MT {field}")
    print(f"p value : {p_value}")
    if p_value <= 0.05:
        print(f"{p_value} <= 0.05, therefore reject H0\n")
    else:
        print(f"{p_value} > 0.05, therefore accept H0\n")
```

#### H0 : Passenger Count Pre Covid Lockdown(Jan-Feb 2020) = Passenger Count Post Covid Lockdown(2 month durations starting June 2020 to present)
#### H1 : Passenger Count Pre Covid Lockdown(Jan-Feb 2020) != Passenger Count Post Covid Lockdown(2 month durations starting June 2020 to present)

```python
dates = [
    ['2020-06-01', '2020-07-31'],
    ['2020-08-01', '2020-09-30'],
    ['2020-10-01', '2020-11-31'],
    ['2020-12-01', '2021-01-31'],
    ['2021-02-01', '2021-03-31'],
    ['2021-04-01', '2021-05-31'],
    ['2021-06-01', '2021-07-31'],
    ['2021-08-01', '2021-09-30'],
    ['2021-10-01', '2021-11-30'],
    ['2021-12-01', '2022-01-31'],
    ['2022-02-01', '2022-03-31']
]

durations = [
    'Jun 2020 - Jul 2020',
    'Aug 2020 - Sep 2020',
    'Oct 2020 - Nov 2020',
    'Dec 2020 - Jan 2021',
    'Feb 2021 - Mar 2021',
    'Apr 2020 - May 2021',
    'Jun 2021 - Jul 2021',
    'Aug 2021 - Sep 2021',
    'Oct 2021 - Nov 2021',
    'Dec 2021 - Jan 2022',
    'Feb 2022 - Mar 2022'
]

ks_values = []
for i in range(len(dates)):
    df_post = df.set_index('Date')
    df_post = df_post.loc[dates[i][0]:dates[i][1],:]
    D1 = df_pre['travel_count'].to_numpy()
    D2 = df_post['travel_count'].to_numpy()
    KS_Statistic = KS_test_2_sample(D1, D2)
    ks_values.append(KS_Statistic)
    print(f"H0: CDF of Pre Covid Passenger Count Data(Jan-Feb 2020) = CDF of Post Covid Passenger Count Data({durations[i]})")
    print(f"H0: CDF of Pre Covid Passenger Count Data(Jan-Feb 2020) != CDF of Post Covid Passenger Count Data({durations[i]})\n")
    print(f"KS Statistic : {KS_Statistic}")
    if KS_Statistic > 0.05:
        print(f"{KS_Statistic} > 0.05, therefore reject H0\n")
    else:
        print(f"{KS_Statistic} <= 0.05, therefore accept H0\n")
        
import matplotlib.pyplot as plt

plt.bar(x=durations, height=ks_values)
plt.xticks(rotation=60)
plt.xlabel("Duration of Post Covid Passenger Count Data")
plt.ylabel("KS Statistic")
plt.show()
```
